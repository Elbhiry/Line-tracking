```python

def getHistogram(img,minper=0.1,region =1):
    
    if region ==1:
       histvalues=np.sum(img,axis=0)
        
    else :
       histvalues=np.sum(img[img.shape[0]//region:,:],axis=0) 
    
        
        
    histvalues=np.sum(img,axis=0)
    maxvalue=np.max(histvalues)
    minvalue=maxvalue*minper
    indexarray=np.where(histvalues >=minvalue)
    basepoint=int(np.average(indexarray))
    
   #if display:
        #imghist = np.zeros((img.shape[0],img.shape[1],3), np.uint8 )
        #for x, intensity in enumerate(histvalues):
        #    cv2.line(imghist,(x,img.shape[0]),(x,img.shape[0]-intensity//255//region),(255,0,255),1)
       #     cv2.circle(imghist,(basepoint,img.shape[0]),20,(0,255,255),cv2.FILLED)    
         #return basepoint
    #print(histvalues)  
    #print(indexarray)
    return basepoint
```

```python

cap = cv2.VideoCapture(0)
while True:                                                      # Load the input frame and write output frame.
    ret, img = cap.read()   
    if not ret:                                                  # Check if frame present otherwise he break the while loop
        break
    ## Press the enter key to exit the video
    if cv2.waitKey(1) == 13: #13 Enter Key
        break
        
#     lower_bound = np.array([0, 0, 0])
#     upper_bound = np.array([180,40,255])
#     hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
#     #masking the image using inRange() function
#     mask = cv2.inRange(hsv, lower_bound, upper_bound)
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    threshold = 200
    binary=cv2.inRange(gray,threshold,255)
    #imagemask = cv2.inRange(hsv, lower_bound, upper_bound)
    #displaying the resulting masked imag
    #mask = cv2.bitwise_or(mask1, mask2)
    #target = cv2.bitwise_and(img,img, mask=mask)

    #cv2.imwrite("target.png", target)
    #cv2.imwrite("RGGB Image.png", mask)
    #kmg=cv2.resize(lmg,(480,240))
    
#     middlepoint=getHistogram(target,minper=0.5,region=4)
#     curveaveragepoint=getHistogram(target,minper=0.9)
#     curveraw =curveaveragepoint- middlepoint
    middlepoint=getHistogram(binary,minper=0.5,region=4)
    curveaveragepoint=getHistogram(binary,minper=0.9)
    curveraw =curveaveragepoint- middlepoint
    #important could make error
    
    command = "N/A"
    #print(curveraw)
    if curveraw>5 :
        #write here ya saad an almotor goes right
        command  = "right"
    elif curveraw<-5:
        #al3ks an almotor ymove left
        command="left"
    else:
        #none willbe done for motor indicating that its a straight line
        command="straight"
        
    #w hena almotor ale haefdl mdy speed ll3rbya 3shan tmshy
    command = command + str(curveraw)
    cv2.putText(binary, command, (10, 20), fontFace=cv2.FONT_HERSHEY_TRIPLEX, fontScale=0.5,color=(255,255,255),thickness=1)
    cv2.imshow('lane detection image', binary)
```
